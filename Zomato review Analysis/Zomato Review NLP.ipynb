NLP

Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing DataSet

```python
data_set = pd.read_csv('Zomato Review Dataset.tsv' , delimiter='\t',quoting = 3)
```

  Cleaning the Data Set

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

corpus = []
for i in range(0,1000):
  zomato_rating = re.sub('[^a-zA-Z]',' ',data_set['Review'][i])
  zomato_rating = zomato_rating.lower()
  zomato_rating = zomato_rating.split()
  pa = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  zomato_rating = [pa.stem(word) for word in zomato_rating if not word in set(all_stopwords)]
  zomato_rating = ' '.join(zomato_rating)
  corpus.append(zomato_rating)
```

Bag Of Words Creation

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
y = data_set.iloc[: , -1].values
```

Training and Testing of Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 21)
```

Training Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
```

Predicting Values on Test Set

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

Making Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
